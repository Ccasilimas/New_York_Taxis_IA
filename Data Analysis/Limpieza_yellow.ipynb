```python
import pandas as pd
from shapely import wkt
from shapely.geometry import Polygon, MultiPolygon
import folium

# Definir la ruta del archivo CSV de zonas de ubicación
zone_file_path = 'taxi_zones.csv'  # Ajusta la ruta según tu archivo

# Leer el dataset de zonas de ubicación de taxis
df_zones = pd.read_csv(zone_file_path)

# Convertir la columna 'the_geom' de WKT a objetos geométricos
df_zones['geometry'] = df_zones['the_geom'].apply(wkt.loads)
```

```python
# Extraer el primer multipolygon del dataframe
first_multipolygon = df_zones.iloc[0]['geometry']
```

```python
# Crear un mapa centrado en Nueva York
ny_map = folium.Map(location=[40.7128, -74.0060], zoom_start=12)

# Añadir el primer multipolygon al mapa
if isinstance(first_multipolygon, MultiPolygon):
    for polygon in first_multipolygon.geoms:
        coords = [(point[1], point[0]) for point in polygon.exterior.coords]
        folium.Polygon(locations=coords, color='blue', fill=True, fill_opacity=0.4).add_to(ny_map)
elif isinstance(first_multipolygon, Polygon):
    coords = [(point[1], point[0]) for point in first_multipolygon.exterior.coords]
    folium.Polygon(locations=coords, color='blue', fill=True, fill_opacity=0.4).add_to(ny_map)

# Guardar el mapa en un archivo HTML
ny_map.save('nyc_first_multipolygon_map.html')
ny_map
```
