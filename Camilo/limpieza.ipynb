```python
import pandas as pd

ruta_archivo_medallions = r'C:\\Users\\kcasi\\Downloads\\Medallion__Vehicles_-_Authorized_20241107.csv'
chunk_size = 10000  # Nmero de filas por trozo

# Inicializar contadores y acumuladores
null_counts = None
unique_medallion_type_counts = None

# Lista de columnas irrelevantes a eliminar
columnas_a_borrar = [
    'Name', 'Expiration Date', 'Current Status', 'DMV License Plate Number',
    'Medallion Type', 'Agent Number', 'Agent Telephone Number', 
    'Agent Website Address', 'Agent Address', 'Last Date Updated', 
    'Last Time Updated'
]

# Leer el archivo CSV en trozos
chunks = pd.read_csv(ruta_archivo_medallions, chunksize=chunk_size)

# Procesar los chunks y escribirlos en el archivo CSV
for i, chunk in enumerate(chunks):
    print(f"Procesando chunk {i + 1}")
    
    # Eliminar columnas irrelevantes del chunk
    chunk = chunk.drop(columns=columnas_a_borrar, errors='ignore')
    
    # Contar valores nulos en cada columna
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    
    # Contar valores nicos en 'Medallion Type'
    if 'Medallion Type' in chunk.columns:
        if unique_medallion_type_counts is None:
            unique_medallion_type_counts = chunk['Medallion Type'].value_counts()
        else:
            unique_medallion_type_counts = unique_medallion_type_counts.add(chunk['Medallion Type'].value_counts(), fill_value=0)
    
    # Escribir el chunk en el archivo CSV
    if i == 0:
        chunk.to_csv('Medallion.csv', index=False, mode='w')
    else:
        chunk.to_csv('Medallion.csv', index=False, mode='a', header=False)

# Mostrar resultados
print("\nConteo de valores nulos por columna:")
print(null_counts)

if unique_medallion_type_counts is not None:
    print("\nConteo de valores nicos en 'Medallion Type':")
    print(unique_medallion_type_counts)
```
